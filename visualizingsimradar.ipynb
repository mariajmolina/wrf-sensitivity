```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import datetime
import os
from wrf import to_np, getvar, get_basemap, latlon_coords, interplevel, smooth2d
from netCDF4 import Dataset
from metpy.plots import colortables
#import matplotlib as mpl
import pandas as pd
from datetime import timedelta
from itertools import product
import geopandas as gpd
import shapely
from descartes import PolygonPatch
from matplotlib.collections import PatchCollection
from mpl_toolkits import basemap
import matplotlib.patches as mpatches
```

```python
def plot(directory, time="2017-01-21_21:00:00"):

    fig = plt.figure(figsize=(10.,8.))
    ncfile2_wrf_d02 = Dataset(f"/glade/scratch/molina/WRF_HYSPLIT_proj/{directory}/wrfout_d01_{time}")
    
    while True:
        dbz2_wrf_d02 = np.array([0])
        if len(dbz2_wrf_d02) == 1:
            try: 
                dbz2_wrf_d02 = getvar(ncfile2_wrf_d02, "dbz")
            except ValueError:
                continue
        if len(dbz2_wrf_d02) != 1:
            break

    lats2, lons2 = latlon_coords(dbz2_wrf_d02)

    bm = get_basemap(dbz2_wrf_d02)

    x, y = bm(to_np(lons2), to_np(lats2))

    wspd_contours = bm.contourf(x, y, to_np(dbz2_wrf_d02)[0,:,:], 
                                levels=[0,10,20,30,40,50,60,70,80],
                                cmap=colortables.get_colortable('NWSReflectivity'), 
                                vmin=0, vmax=80)

    bm.drawcoastlines(linewidth=0.75, zorder=10)
    bm.drawstates(zorder=10)
    bm.drawcountries(zorder=10)
    bm.drawparallels(np.arange(int(20),int(50),5), labels=[1,0,0,0], linewidth=0.0, fontsize=11)
    return plt.show()
```

```python
def plot2(directory='wrf4km_ens_2'):

    
    
    import pandas as pd
    rng = pd.date_range(pd.Timestamp("2017-01-21 09:00"), periods=49, freq='h')
    
    for num, r in enumerate(rng.strftime("%Y-%m-%d_%H:00:00")):
        
        plt.close()
        
        fig = plt.figure(figsize=(10.,8.))
        ncfile2_wrf_d02 = Dataset(f"/glade/scratch/molina/WRF_HYSPLIT_proj/{directory}/wrfout_d01_{r}")

        while True:
            dbz2_wrf_d02 = np.array([0])
            if len(dbz2_wrf_d02) == 1:
                try: 
                    dbz2_wrf_d02 = getvar(ncfile2_wrf_d02, "dbz")
                except ValueError:
                    continue
            if len(dbz2_wrf_d02) != 1:
                break

        lats2, lons2 = latlon_coords(dbz2_wrf_d02)

        bm = get_basemap(dbz2_wrf_d02)

        x, y = bm(to_np(lons2), to_np(lats2))

        bm.contourf(x, y, to_np(dbz2_wrf_d02)[0,:,:], 
                                    levels=[0,10,20,30,40,50,60,70,80],
                                    cmap=colortables.get_colortable('NWSReflectivity'), 
                                    vmin=0, vmax=80)

        bm.drawcoastlines(linewidth=0.75, zorder=10)
        bm.drawstates(zorder=10)
        bm.drawcountries(zorder=10)
        bm.drawparallels(np.arange(int(20),int(50),5), labels=[1,0,0,0], linewidth=0.0, fontsize=11)
        
        plt.savefig(f"/glade/scratch/molina/WRF_HYSPLIT_proj/{directory}/pic_{num}.png", bbox_inches='tight')
        
    return
```

```python
plot2()
```

```python
import subprocess

def grid2gif(image_str, output_gif):
    str1 = 'convert -delay 8 -loop 0 ' + image_str  + ' ' + output_gif
    subprocess.call(str1, shell=True)
```
