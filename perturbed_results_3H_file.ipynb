```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

#the control run
wrf_control_orig_0 = 'wrf4km_ens_2'

wrf_gombasn_cool_1 = 'wrf4km_sst_gombasin_cool_1'
wrf_gombasn_cool_2 = 'wrf4km_sst_gombasin_cool_2'
wrf_gomplum_cool_1 = 'wrf4km_sst_gomplume_cool_1'
wrf_gomplum_cool_2 = 'wrf4km_sst_gomplume_cool_2'
wrf_murgion_cool_1 = 'wrf4km_sst_muregion_cool_1'
wrf_murgion_cool_2 = 'wrf4km_sst_muregion_cool_2'

wrf_gombasn_warm_1 = 'wrf4km_sst_gombasin_warm_1'
wrf_gombasn_warm_2 = 'wrf4km_sst_gombasin_warm_2'
wrf_gomplum_warm_1 = 'wrf4km_sst_gomplume_warm_1'
wrf_gomplum_warm_2 = 'wrf4km_sst_gomplume_warm_2'
wrf_murgion_warm_1 = 'wrf4km_sst_muregion_warm_1'
wrf_murgion_warm_2 = 'wrf4km_sst_muregion_warm_2'

UH_thresholds = np.array([100,100,100,80,80,80,60,60,60,40,40,40])
dbz_thresholds = np.array([50,40,30,50,40,30,50,40,30,50,40,30])

mask_data = xr.open_dataset('/glade/work/molina/DATA/usstates_shapefiles/conus_mask_res_80.nc')


def percent_change(data1,data2,data_mask):
   
    return np.divide((data1.UHdbz.where(data_mask.conus.values, 0).sum()-
                       data2.UHdbz_nomask.where(data_mask.conus.values, 0).sum()),
                       data2.UHdbz_nomask.where(data_mask.conus.values, 0).sum())*100


def percent_change_scalar(data1,data2):
   
    return np.divide((data1-data2),data2)*100
```

```python
thegombasin = np.zeros((len(UH_thresholds),5))
thegomplume = np.zeros((len(UH_thresholds),5))
themuregion = np.zeros((len(UH_thresholds),5))
thegombasin_perc = np.zeros((len(UH_thresholds),5))
thegomplume_perc = np.zeros((len(UH_thresholds),5))
themuregion_perc = np.zeros((len(UH_thresholds),5))

thresh = []

for nume, (UH_threshold, dbz_threshold) in enumerate(zip(UH_thresholds, dbz_thresholds)):

    
    thresh.append('UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold))
    
    control_orig_0=xr.open_dataset('/glade/work/molina/DATA/jan2017_synoptic/'+wrf_control_orig_0+'/newver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_control_orig_0 = np.where(mask_data.conus.values,np.sum(np.where(control_orig_0.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)

    gombasn_cool_1=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_gombasn_cool_1+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_gombasn_cool_1 = np.where(mask_data.conus.values,np.sum(np.where(gombasn_cool_1.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)

    gombasn_cool_2=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_gombasn_cool_2+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_gombasn_cool_2 = np.where(mask_data.conus.values,np.sum(np.where(gombasn_cool_2.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)
    
    gomplum_cool_1=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_gomplum_cool_1+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_gomplum_cool_1 = np.where(mask_data.conus.values,np.sum(np.where(gomplum_cool_1.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)
    
    gomplum_cool_2=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_gomplum_cool_2+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_gomplum_cool_2 = np.where(mask_data.conus.values,np.sum(np.where(gomplum_cool_2.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)
    
    murgion_cool_1=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_murgion_cool_1+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_murgion_cool_1 = np.where(mask_data.conus.values,np.sum(np.where(murgion_cool_1.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)
    
    murgion_cool_2=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_murgion_cool_2+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_murgion_cool_2 = np.where(mask_data.conus.values,np.sum(np.where(murgion_cool_2.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)

    gombasn_warm_1=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_gombasn_warm_1+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_gombasn_warm_1 = np.where(mask_data.conus.values,np.sum(np.where(gombasn_warm_1.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)
    
    gombasn_warm_2=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_gombasn_warm_2+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_gombasn_warm_2 = np.where(mask_data.conus.values,np.sum(np.where(gombasn_warm_2.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)
    
    gomplum_warm_1=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_gomplum_warm_1+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_gomplum_warm_1 = np.where(mask_data.conus.values,np.sum(np.where(gomplum_warm_1.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)
    
    gomplum_warm_2=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_gomplum_warm_2+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_gomplum_warm_2 = np.where(mask_data.conus.values,np.sum(np.where(gomplum_warm_2.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)
    
    murgion_warm_1=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_murgion_warm_1+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_murgion_warm_1 = np.where(mask_data.conus.values,np.sum(np.where(murgion_warm_1.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)
    
    murgion_warm_2=xr.open_dataset('/glade/scratch/molina/WRF_HYSPLIT_proj/'+wrf_murgion_warm_2+'/ver_grid_80_UH_'+str(UH_threshold)+'_Z_'+str(dbz_threshold)+'.nc')
    new_murgion_warm_2 = np.where(mask_data.conus.values,np.sum(np.where(murgion_warm_2.UHdbz_dates.resample(time='3H').sum('time')>0,1,0),axis=0),0)
    

    gombasin = np.array([new_gombasn_cool_2.sum(), 
                         new_gombasn_cool_1.sum(), 
                         new_control_orig_0.sum(), 
                         new_gombasn_warm_1.sum(), 
                         new_gombasn_warm_2.sum()])

    gomplume = np.array([new_gomplum_cool_2.sum(), 
                         new_gomplum_cool_1.sum(), 
                         new_control_orig_0.sum(), 
                         new_gomplum_warm_1.sum(), 
                         new_gomplum_warm_2.sum()])

    muregion = np.array([new_murgion_cool_2.sum(), 
                         new_murgion_cool_1.sum(), 
                         new_control_orig_0.sum(), 
                         new_murgion_warm_1.sum(), 
                         new_murgion_warm_2.sum()])
    
    thegombasin[nume] = gombasin
    thegomplume[nume] = gomplume
    themuregion[nume] = muregion
    
    
    perc_gombasin = np.array([percent_change_scalar(new_gombasn_cool_2.sum(),new_control_orig_0.sum()),
                              percent_change_scalar(new_gombasn_cool_1.sum(),new_control_orig_0.sum()),
                              0.,
                              percent_change_scalar(new_gombasn_warm_1.sum(),new_control_orig_0.sum()),
                              percent_change_scalar(new_gombasn_warm_2.sum(),new_control_orig_0.sum())])
    
    perc_gomplume = np.array([percent_change_scalar(new_gomplum_cool_2.sum(),new_control_orig_0.sum()),
                              percent_change_scalar(new_gomplum_cool_1.sum(),new_control_orig_0.sum()),
                              0.,
                              percent_change_scalar(new_gomplum_warm_1.sum(),new_control_orig_0.sum()),
                              percent_change_scalar(new_gomplum_warm_2.sum(),new_control_orig_0.sum())])
    
    perc_muregion = np.array([percent_change_scalar(new_murgion_cool_2.sum(),new_control_orig_0.sum()),
                              percent_change_scalar(new_murgion_cool_1.sum(),new_control_orig_0.sum()),
                              0.,
                              percent_change_scalar(new_murgion_warm_1.sum(),new_control_orig_0.sum()),
                              percent_change_scalar(new_murgion_warm_2.sum(),new_control_orig_0.sum())])
    
    
    thegombasin_perc[nume] = perc_gombasin
    thegomplume_perc[nume] = perc_gomplume
    themuregion_perc[nume] = perc_muregion
```

```python
bleh = xr.Dataset({'gombasin_sum':(['thresholds','x'],thegombasin),
                   'gomplume_sum':(['thresholds','x'],thegomplume),
                   'muregion_sum':(['thresholds','x'],themuregion),
                   'gombasin_per':(['thresholds','x'],thegombasin_perc),
                   'gomplume_per':(['thresholds','x'],thegomplume_perc),
                   'muregion_per':(['thresholds','x'],themuregion_perc),},
                   coords={'threshold':(['thresholds'],thresh),},
                   attrs={'File Author':'Maria J. Molina'})
```
